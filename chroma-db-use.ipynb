```python
import PyPDF2  # Or another PDF extraction library
from nltk.corpus import stopwords  # For stop word removal
from nltk.stem import PorterStemmer  # For stemming

def extract_text(pdf_path):
    pdf_file = open(pdf_path, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    text = ""
    for page in pdf_reader.pages:
        text += page.extract_text()
    pdf_file.close()
    return text

def preprocess_text(text):
    # Remove stop words, perform stemming/lemmatization
    stop_words = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    preprocessed_text = [stemmer.stem(word) for word in text.lower().split() if word not in stop_words]
    return preprocessed_text

def generate_embedding(text, model_name="sentence-transformers/all-mpnet-base-v2"):
    # Replace with your preferred text embedding model
    from sentence_transformers import SentenceTransformer
    model = SentenceTransformer(model_name)
    embedding = model.encode(text)  # Generates a numerical vector
    return embedding

# Example usage
pdf_path = "/kaggle/input/handbook/fitness-handbook.pdf"
extracted_text = extract_text(pdf_path)
preprocessed_text = preprocess_text(extracted_text)
embedding = generate_embedding(preprocessed_text)

# You'll need to upload the embedding to Chromma Search manually
```
